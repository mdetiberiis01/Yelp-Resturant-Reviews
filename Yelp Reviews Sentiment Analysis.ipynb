```python
import pandas as pd
import numpy as np
import requests
import json
from bs4 import BeautifulSoup
from yelp.client import Client
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#This will be the URL of the Establishment you want to analyse 

R = requests.get('https://www.yelp.com/biz/chipotle-mexican-grill-mount-kisco?osq=Restaurants')
```

```python
R.text
```

```python
soup = BeautifulSoup(R.text, 'html.parser')

```

```python
results = soup.findAll(class_='lemon--span__373c0__3997G raw__373c0__3rKqk', attrs={'lang':'en'})
reviews = [result.text for result in results]
```

```python
for review in reviews:
    print(review,'\n')
 
```

```python
#Analysing the Data 

import pandas as pd
import numpy as np
```

```python
#Create our Data Frame

df = pd.DataFrame(np.array(reviews), columns = ['review'])
```

```python
#Review our Data Frame

df
```

```python
#Gather the word length of each review

df['word_count'] = df['review'].apply(lambda x: len(x.split()))
```

```python
#Gather the character count of each review

df['char_count'] = df['review'].apply(lambda x: len(x))
```

```python
def average_words(x):
    words = x.split() #to give the # of words
    return sum(len(word)for word in words) / len(words)
```

```python
df['average_word_length'] = df['review'].apply(lambda x: average_words(x))
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')

len(stop_words)
```

```python
#How many stop words in each review 

df['stopword_count'] = df['review'].apply(lambda x: len([word for word in x.split() if word in stop_words]))
```

```python
df['stopword_percentage'] = df['stopword_count']/df['word_count']
```

```python
#bc i added a column by accident 

#this is how you remove a column
#df = df.drop(columns=['stop_word_count'])
```

```python
df.sort_values(by = 'stopword_percentage')
```

```python
#Standard metrics

df.describe()
```

```python
#Data Cleaning 

#looping through each word which is x
#splititng x givng us every word
#every word we get back were lowering

df['lowercase'] = df['review'].apply(lambda x: " ".join(word.lower() for word in x.split()))
```

```python
#removing puncuations

df['punctuation'] = df['lowercase'].str.replace('[^\w\s]', '')
```

```python
df['stopwords'] = df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
#an array of every single word in all the reviews

#pd.Series(" ".join(df['stopwords']).split()).value_counts()[:30]
```

```python
#other words that dont add value

other_stopwords = ['us','get','also','would','end','hill']
```

```python
df['clean_review'] = df['stopwords'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stopwords))
```

```python
#an array of every single word in all the reviews

pd.Series(" ".join(df['clean_review']).split()).value_counts()[:30]
```

```python
df.head()
```

```python
#Lemmatization
#Basically changing word like resturants to resturant(words to word) bc really they mean the same thing 

import textblob

from textblob import Word

from nltk.stem import WordNetLemmatizer 
  
lemmatizer = WordNetLemmatizer() 
```

```python
df['lemmatized'] = df['clean_review'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
#Sentiment Analysis

from textblob import TextBlob
```

```python
"""When doing sentiment the two metrics you get back are #ploarity and
#subjectivity. polarity gives you from -1 to 1 a metric of how positive 
or negative a review is. Subjectivity is a measure between 0 - 1 of how 
much the text is based on factual information or just general opinion"""
```

```python
df['polarity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
#clean up columns we dont need now

df.drop(['lowercase','punctuation','stopwords','clean_review', 'lemmatized'], axis = 1, inplace = True)

```

```python
df = df.sort_values(by = 'polarity', ascending = False)
```
